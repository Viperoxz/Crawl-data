```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cereal_df = pd.read_csv('cereal.csv')
cereal_df
```

```python
cereal_df.drop(['shelf', 'weight', 'cups', 'rating'], axis=1, inplace=True)
cereal_df
```

```python
calo = cereal_df.calories.mean()
pro = cereal_df.protein.mean()
pro/calo
```

```python
cereal_corr = cereal_df.corr()
cereal_corr
```

```python
maskz = np.ones_like(cereal_corr, dtype=bool)
maskz
```

```python
maskz = np.triu(maskz)
maskz
```

```python
sns.heatmap(data=cereal_corr, cmap='Blues', mask=maskz)
```

```python
adjusted_mask = maskz[1:, :-1]
adjusted_mask
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
adjusted_cereal_corr
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
c_map = sns.diverging_palette(250, 0, 80, 60, as_cmap=True, center='light')
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask, cmap=c_map, annot=True, fmt='.2f', vmin=-1, vmax=1, linecolor='white', linewidths=0.5)
yticks = [i.upper() for i in adjusted_cereal_corr .index] 
xticks = [i.upper() for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks, rotation=0);
ax.set_xticklabels(xticks);
ax.set_title('CORRELATION MATRIX SAMPLED CEREALS COMPOSITION', fontsize=18);
```
