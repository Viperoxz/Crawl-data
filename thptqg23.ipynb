```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
diemthi = []
```

```python
def readFile(filename, sheetname, skipr):
    return pd.read_excel(filename, sheet_name=sheetname, skiprows=skipr)
```

```python
for stu in range(47000001, 47012901):
    scrap_url = 'https://dantri.com.vn/thpt/1/0/99/' + str(stu) + '/2023/0.2/search-gradle.htm'
    payload, header = {}, {}
    response = requests.request('GET', scrap_url, headers=header, data=payload)
    info = response.json()['student']
    diem = '{}    {}    {}    {}    {}    {}    {}    {}    {}    {}    {}    {}'.format(info['sbd'],
    info['toan'], info['van'], info['ngoaiNgu'], info['vatLy'], info['hoaHoc'], info['sinhHoc'], info['diemTBTuNhien'], info['lichSu'], info['diaLy'], info['gdcd'], info['diemTBXaHoi'])
    diemthi.append(diem)


```

```python
df = df.astype({'SBD' : 'int64', 'Toan' : 'float64', 'Van' : 'float64', 'Ngoai ngu' : 'float64',
                 'Ly' : 'float64', 'Hoa' : 'float64', 'Sinh' : 'float64', 'TBKHTN' : 'float64', 'Su' : 'float64', 'Dia' : 'float64', 'GDCD' : 'float64', 'TBKHXH' : 'float64'})
```

```python
df['A0'] = df['Toan'] + df['Ly'] + df['Hoa']
df['A1'] = df['Toan'] + df['Ly'] + df['Ngoai ngu']
df['B0'] = df['Toan'] + df['Sinh'] + df['Hoa']
df['D1'] = df['Toan'] + df['Ngoai ngu'] + df['Van']
```

```python
df_a0 = df.sort_values(by=['A0'], ascending=False)
df_a0['Rank'] = df_a0['A0'].rank(method='min', ascending=False)
df_a0['Rank'] = df_a0['Rank'].astype('int64')
df_a0
```

```python
df_a1 = df.sort_values(by=['A1'], ascending=False)
df_a1['Rank'] = df_a1['A1'].rank(method='min', ascending=False)
df_a1['Rank'] = df_a1['Rank'].astype('int64')
df_a1
```

```python
df_b0 = df.sort_values(by=['B0'], ascending=False)
df_b0['Rank'] = df_b0['B0'].rank(method='min', ascending=False)
df_b0['Rank'] = df_b0['Rank'].astype('int64')
df_b0
```

```python
df_d1 = df.sort_values(by=['D1'], ascending=False)
df_d1['Rank'] = df_d1['D1'].rank(method='min', ascending=False)
df_d1['Rank'] = df_d1['Rank'].astype('int64')
df_d1
```
